# Matrices and Vectors

---

Trong phân tích dữ liệu nhiều biến, ta quan tâm đến việc phân tích đồng thời nhiều biến phụ thuộc *(dependent variables)*. Các biến phụ thuộc này sẽ được biểu diễn dưới dạng các ma trận và vector.

Tập dữ liệu với $n$ quan sát *(observations)* và $p$ biến *(variables)* sẽ được biểu diễn dưới dạng ma trận như sau:

\begin{equation}
X =
\begin{bmatrix}
   x_{11} & x_{12} & \dots & x_{1p} \\
   x_{21} & x_{22} & \dots & x_{2p} \\
   \vdots & \vdots & \ddots & \vdots \\
   x_{n1} & x_{n2} & \dots & x_{np} \\
\end{bmatrix}
\end{equation}

Trong đó, biến $x_{ij}$ là giá trị của biến thứ j cho quan sát thứ i.

Các vector hàng $R_i = \begin{bmatrix}
   x_{i1} & x_{i2} & \dots & x_{ip} \\
\end{bmatrix}$ đại diện cho một quan sát

và các vector cột $C_i = \begin{bmatrix}
   x_{1i} \\
   x_{2i} \\
   \vdots \\
   x_{ni} \\
\end{bmatrix}$ đại diện cho một biến

# Descriptive Statistic (Thống kê mô tả)

---

**1.   Trung bình (Mean)**

Giá trị trung bình tổng thể *(population mean)* của từng biến thể hiện xu hướng trung tâm *(central tendency)* của tổng thể đó, kí hiệu là $\mu_k = E(x_k)$

Giá trị trung bình tổng thể có thể được ước lượng bởi giá trị trung bình mẫu *(sample mean)*:

\begin{equation}
\bar{x_k} = \frac{1}{n}\displaystyle\sum_{i=1}^{n}x_{ik}, k=1, 2,..., p
\end{equation}

Ta có thể ước lượng vector trung bình tổng thể $\mu = \begin{bmatrix}
   \mu_1 \\
   \mu_2 \\
   \vdots \\
   \mu_p
\end{bmatrix}$ bởi vector trung bình mẫu $\bar{x} = \begin{bmatrix}
   \bar{x_1} \\
   \bar{x_2} \\
   \vdots \\
   \bar{x_p}
\end{bmatrix}$

**2.   Phương sai & Độ lệch chuẩn (Variance & Standard deviation)**

Phương sai *(variance)* cho biết mức độ phân tán *(dispersion)* của các giá trị trong một biến:

\begin{equation}
s_{kk} = \frac{1}{n}\displaystyle\sum_{i=1}^{n}(x_{ik}-\bar{x_k})^2, k=1, 2,..., p
\end{equation}

Căn bậc hai của phương sai được gọi là độ lệch chuẩn *(standard deviation)*:

\begin{equation}
s_k = \sqrt{s_{kk}}
\end{equation}

**3.   Hiệp phương sai & Hệ số tương quan (Covariance & Correlation)**

Hiệp phương sai *(covariance)* cho biết sự tương quan giữa mỗi biến với các biến khác:

\begin{equation}
s_{jk} = \frac{1}{n}\displaystyle\sum_{i=1}^{n}(x_{ij} - \bar{x_j})(x_{ik} - \bar{x_k})
\end{equation}

Ma trận hiệp phương sai:

\begin{equation}
S =
\begin{bmatrix}
   s_{11} & s_{12} & \dots & s_{1p} \\
   s_{21} & s_{22} & \dots & s_{2p} \\
   \vdots & \vdots & \ddots & \vdots \\
   s_{p1} & s_{p2} & \dots & s_{pp} \\
\end{bmatrix}
\end{equation}

Hệ số tương quan *(Correlation - Pearson's correlation coefficient)* đo lường mức độ tương quan tuyến tính giữa hai biến. Khác với hiệp phương sai, hệ số tương quan đã được chuẩn hoá và không bị phụ thuộc vào đơn vị của các biến $\to$ sử dụng để so sánh mức tương quan của hai biến này với mức tương quan của hai biến kia:

\begin{equation}
r_{jk} = \frac{s_{jk}}{\sqrt{s_{jj}}\sqrt{s_{kk}}} = \frac{\displaystyle\sum_{i=1}^{n}(x_{ij} - \bar{x_j})(x_{ik} - \bar{x_k})}{\sqrt{\displaystyle\sum_{i=1}^{n}(x_{ij}-\bar{x_j})^2}\sqrt{\displaystyle\sum_{i=1}^{n}(x_{ik}-\bar{x_k})^2}}
\end{equation}

Ma trận tương quan:

\begin{equation}
R =
\begin{bmatrix}
   1 & r_{12} & \dots & r_{1p} \\
   r_{21} & 1 & \dots & r_{2p} \\
   \vdots & \vdots & \ddots & \vdots \\
   r_{p1} & r_{p2} & \dots & 1 \\
\end{bmatrix}
\end{equation}

# Distance (Khoảng cách)

---

*   **Euclidean distance (Khoảng cách Euclidean)**

Khoảng cách Euclidean từ điểm $P = (x_1, x_2,...,x_p)$ tới gốc toạ độ O:

\begin{equation}
d(O, P) = \sqrt{x_1^2 + x_2^2 + ... + x_p^2}
\end{equation}

Khoảng cách Euclidean giữa hai điểm $P = (x_1, x_2,...,x_p)$ và $Q = (y_1, y_2,...,y_p)$:

\begin{equation}
d(P, Q) = \sqrt{(x_1-y_1)^2 + (x_2-y_2)^2 + ... + (x_p-y_p)^2}
\end{equation}


*   **Statistical distance (Khoảng cách thống kê)**

Trong công thức khoảng cách Euclidean, có thể thấy mức độ đóng góp của các trục toạ độ là bằng nhau. Tuy nhiên, các trục toạ độ đại diện cho các biến thường sẽ có phương sai khác nhau, dẫn đến việc cần đặt trọng số tương ứng với phương sai của mỗi biến, cụ thể hơn, các biến có phương sai lớn hơn sẽ được đặt trọng số ít hơn so với những biến có phương sai bé hơn $\to$ khoảng cách thống kê

Khoảng cách thống kê từ điểm $P = (x_1, x_2,...,x_p)$ tới gốc toạ độ O:

\begin{equation}
d(O, P) = \sqrt{\frac{x_1^2}{s_{11}} + \frac{x_2^2}{s_{22}} + ... + \frac{x_p^2}{s_{pp}}}
\end{equation}

Khoảng cách thống kê giữa hai điểm $P = (x_1, x_2,...,x_p)$ và $Q = (y_1, y_2,...,y_p)$:

\begin{equation}
d(P, Q) = \sqrt{\frac{(x_1-y_1)^2}{s_{11}} + \frac{(x_2-y_2)^2}{s_{22}} + ... + \frac{(x_p-y_p)^2}{s_{pp}}}
\end{equation}

Công thức khoảng cách thống kê phía trên vẫn chưa giải quyết được hoàn toàn các vấn đề trong thống kê do đang giả sử hai biến là độc lập với nhau $\to$ Để tính khoảng cách giữa các điểm có ý nghĩa hơn trong trường hợp các biến có tương quan với nhau, ta sẽ tính khoảng cách thống kê giữa các điểm trong một hệ trục toạ độ mới được tạo ra từ việc xoay hệ trục toạ độ một góc $\theta$ mà trong hệ toạ độ đó sự tương quan giữa các biến là lớn nhất. Việc xác định góc $\theta$ sẽ dựa vào eigen vector (vector riêng) của covariance matrix (ma trận hiệp phương sai) của các biến sau khi đã được centralized. Khi đó:

Khoảng cách thống kê từ điểm $P = (x_1, x_2,...,x_p)$ tới gốc toạ độ O:

\begin{equation}
d(O, P) = \sqrt{a_{11}x_1^2 + a_{22}x_2^2 + ... + a_{pp}x_p^2 + 2a_{12}x_1x_2 + 2a_{13}x_1x_3 + ... + 2a_{p-1,p}x_{p-1}x_p}
\end{equation}

Khoảng cách thống kê giữa hai điểm $P = (x_1, x_2,...,x_p)$ và $Q = (y_1, y_2,...,y_p)$:

\begin{equation}
d(P, Q) = \sqrt{a_{11}(x_1-y_1)^2 + a_{22}(x_2-y_2)^2 + ... + a_{pp}(x_p-y_p)^2 + 2a_{12}(x_1-y_1)(x_2-y_2) + 2a_{13}(x_1-y_1)(x_3-y_3) + ... + 2a_{p-1,p}(x_{p-1}-y_{p-1})(x_p-y_p)}
\end{equation}

hay
\begin{equation}
d^2(P, Q) = \begin{bmatrix}
   x_1 - y_1 & x_2 - y_2 & \dots & x_p - y_p \\
\end{bmatrix} \begin{bmatrix}
   a_{11} & a_{12} & \dots & a_{1p} \\
   a_{12} & a_{22} & \dots & a_{2p} \\
   \vdots & \vdots & \ddots & \vdots \\
   a_{1p} & a_{2p} & \dots & a_{pp} \\
\end{bmatrix} \begin{bmatrix}
   x_1 - y_1 \\
   x_2 - y_2 \\
   \vdots \\
   x_p - y_p \\
\end{bmatrix}
\end{equation} (dạng toàn phương)

trong đó, ma trận $\begin{bmatrix}
   a_{11} & a_{12} & \dots & a_{1p} \\
   a_{12} & a_{22} & \dots & a_{2p} \\
   \vdots & \vdots & \ddots & \vdots \\
   a_{1p} & a_{2p} & \dots & a_{pp} \\
\end{bmatrix}$ là ma trận đối xứng xác định dương được xác định từ góc xoay $\theta$

<img src="https://drive.google.com/uc?id=1ywlbJ8MgXr3fttJ90_ZgQg2Ks0n-ZA2X" width="500" alt="statistical distance"/>

Ta có thể thấy khoảng cách Euclidean từ điểm $Q$ đến $P$ sẽ lớn hơn khoảng cách Euclidean từ $Q$ đến gốc toạ độ $O$. Tuy nhiên, khi áp dụng công thức khoảng cách thống kê với hệ trục toạ độ mới, khoảng cách thống kê từ $Q$ đến $P$ sẽ nhỏ hơn khoảng cách thống kê từ $Q$ đến $O$, và điều này là hợp lí vì quan sát hình có thể thấy điểm $P$ có khả năng thuộc cùng một nhóm với $Q$ cao hơn điểm $O$

# Basics of Matrix and Vector Algebra (Đại số Ma trận và Vector)

---

*Recommend:* [*Essence of linear algebra*](https://youtube.com/playlist?list=PLZHQObOWTQDPD3MizzM2xVFitgF8hE_ab&si=I8EbEisReG14u3HN)


**1.   Vector**

Mảng $x$ gồm $n$ số thực $x_1, x_2,... x_n$ được gọi là vector, kí hiệu: $x = \begin{bmatrix}
   x_1 \\
   x_2 \\
   \vdots \\
   x_n
\end{bmatrix}$

Có thể hiểu vector là các đại lượng có độ lớn và chỉ hướng trong không gian. Trong nội dung liên quan đến môn học này, vector trong hệ trục toạ độ là một đoạn thẳng có hướng với đuôi vector luôn nằm ở gốc toạ độ

<img src="https://drive.google.com/uc?id=1kuHNZGFuKReLXHoisLHPs74INw-3L1oR" width="500" alt="vector"/>

<u>Ý nghĩa của các con số</u>: Toạ độ của vector là các con số thể hiện đường đi từ đuôi vector ở gốc toạ độ đến đỉnh của vector.

> Ví dụ: Toạ độ của vector trong hệ toạ độ 2 chiều có số thứ nhất cho biết độ dài và hướng đi theo trục $x$ (dương đi sang phải, âm đi sang trái), số thứ hai cho biết độ dài và hướng đi theo trục $y$ (dương đi lên trên, âm đi xuống dưới)
  <img src="https://drive.google.com/uc?id=1fhT0zK5yoxnyZssHVeXDk6tzBwHKpv38" width="500" alt="vector in 2D coordinate system"/>

<u>Vector chuyển vị</u>: $x^T = \begin{bmatrix}
   x_1 & x_2 & \dots x_n \\
\end{bmatrix}$

<u>Phép toán với vector</u>:

*   *Addition (Phép cộng)*

$\begin{bmatrix}
   x_1 \\
   y_1 \\
\end{bmatrix} + \begin{bmatrix}
   x_2 \\
   y_2 \\
\end{bmatrix} = \begin{bmatrix}
   x_1 + x_2\\
   y_1 + y_2\\
\end{bmatrix}$

(Ở đây có thể hiểu kết quả của phép cộng 2 vector là vector có toạ độ dẫn tới điểm đến cuối cùng sau khi di chuyển lần lượt theo toạ độ của 2 vector đó)

<img src="https://drive.google.com/uc?id=13LjLqQO0jX1aq_akGWvZ8ox5y3b3eecA" width="500" alt="vector addition"/>

*   *Scalar multiplication (Phép nhân với số vô hướng)*

$c\begin{bmatrix}
   x \\
   y \\
\end{bmatrix} = \begin{bmatrix}
   cx\\
   cy\\
\end{bmatrix}$

(Tuỳ vào giá trị của số vô hướng mà scalar multiplication sẽ kéo dài, làm ngắn hay đảo ngược vector, hay nói cách khác là scaling $to$ để chỉ các số vô hướng đó có thể dùng từ 'scalar' thay thế cho từ 'number')
